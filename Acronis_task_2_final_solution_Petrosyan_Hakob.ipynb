```python
import pandas as pd
import numpy as np
from datetime import datetime

data = pd.read_csv('ms_bucket.csv',)
data = data.fillna(0)
timestamp = data["timestamp"] # duplication for graphic 
```

```python
# Timestamp to date
timestamp = np.array(data["timestamp"])
for i in range(len(timestamp)):
    data.loc[i, "timestamp"] = datetime.fromtimestamp(timestamp[i])
```

```python
# Sorting columns
old_columns = np.array(data.columns[2:])
for i in range(0, len(old_columns)):
    old_columns[i] = int(old_columns[i])
old_columns.sort()
for i in range(0, len(old_columns)):
    old_columns[i] = str(old_columns[i])
y = np.array(["timestamp", "Inf"]) 
old_columns = np.append(y, old_columns)
data = data[old_columns]
data
```

```python
arr = np.diff(data.values[:, 2:], axis=1)
arr
```

```python
# After compiling of this code, each element (i = 3...33) in the data.columns
# shows that the process was completed in (data.columns[i]-data.columns[i-1]) milliseconds.
# This is done in order to avoid at least some dependence on columns
y = data.columns
for i in range(2, len(data.columns)-1):
    data[y[i+1]] = arr[:, (i-2):(i-1)]
data
```

```python
# String to int
columns = np.array(data.columns[2:])
for i in range(0, len(columns)):
    columns[i] = int(columns[i])
columns
```

```python
# Getting time intervals for each column
dist = [1]
for i in range(len(columns)-1):
    dist.append(columns[i+1] - columns[i])
dist = np.array(dist)
dist
```

```python
x = np.array([[0 for i in range(32)] for i in range(8352)])
for k in range(8352):
    x[k]= np.array(data.iloc[k][2:]) / dist # a kind of normalization of each element to get more or less normal weights


mean = []
vector = np.array(data.columns[2:])
for i in range(0, len(vector)):
    vector[i] = int(vector[i])
vector = vector.reshape(1, 32)


for k in range(8352):
    mean.append(vector * np.array(x[k]))
    mean[k] = mean[k].sum()
    if mean[k] != 0:
        mean[k] = mean[k] / data["Inf"][k]
mean = np.array(mean)
```

```python
mean
```

```python
std = []
for k in range(8352):
    st = ((vector - mean[k])**2) * np.array(x[k])
    std.append(st.sum())
    if std[k] != 0:
        std[k] = std[k] / data["Inf"][k]
std= np.array(std)
```

```python
std
```
